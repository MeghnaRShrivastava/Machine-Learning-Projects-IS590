```python
#Removing the unnecessary characters from the tweets 
import os
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
import collections

import tweepy as tw
import nltk

import re
import networkx

import warnings
warnings.filterwarnings("ignore")

sns.set(font_scale=1.5)
sns.set_style("darkgrid")
```

```python
#Cleaning the tweets using string command
import string
string.punctuation
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

master_table['Tweet'] = master_table['Tweet'].apply(lambda x: remove_punct(x))
master_table
```

```python
master_table = master_table.groupby(['FileName'])['Tweet'].apply(','.join).reset_index()
master_table
```

```python
#Perform Tokenization
def tokenization(text):
    text = re.split('\W+', text)
    return text

master_table['Tweet'] = master_table['Tweet'].apply(lambda x: tokenization(x.lower()))
master_table
```

```python
#remove stopwords
from nltk.corpus import stopwords
```

```python
stopword = nltk.corpus.stopwords.words('english')
stopword.extend(['yr', 'year', 'woman', 'man', 'girl','boy','one', 'two', 'sixteen', 'yearold', 'fu', 'weeks', 'week',
                 'treatment', 'associated', 'patients', 'may','day', 'case','old','got'])
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
master_table['Tweet'] = master_table['Tweet'].apply(lambda x: remove_stopwords(x))
master_table
```

```python
import nltk
nltk.download('wordnet')
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

master_table['Tweet'] = master_table['Tweet'].apply(lambda x: lemmatizer(x))
master_table
```

```python
from collections import Counter
master_table['Words'] = ""

#counting frequencies
i = 0
for row in master_table['Tweet']:
    master_table['Words'][i] = str(Counter(row).most_common(10))
    i+=1

master_table
```
